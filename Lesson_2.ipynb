### Тема “Множество”
Task№1 Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.<br>
Task№2 Выполнить задание 1 на языке Python


```python
a = {1,2,3}
b = {1,4,6}
c = {2,4,8}
print(f"a | b = {a | b} - a или b, остались только уникальные значения, взятых из обоих множеств")
print(f"a & b = {a & b} - а и b, остались только уникальные значения, входящие одновременно в оба множества ")
print(f"a - b = {a - b} - разница между a и b(остаются только те элементы a, которых нет в b)")
print(f"b - a = {b - a} - разница между b и a(остаются только те элементы b, которых нет в a)")
print(f"a ^ b = {a ^ b} - элементы которые есть либо в а, либо в b, но не в обоих сразу") 
print(f"a | c = {a | c}")
print(f"a & c = {a & c}")
print(f"a - c = {a - c}")
print(f"c - a = {c - a}")
print(f"a ^ c = {a ^ c}")
print(f"c | b = {c | b}")
print(f"c & b = {c & b}")
print(f"c - b = {c - b}")
print(f"b - c = {b - c}")
print(f"c ^ b = {c ^ b}")
```
