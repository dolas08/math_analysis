Последовательность a<sub>n</sub> = 2<sup>n</sup>-n<br>
Имеет вид 2<sup>1</sup>- 1, 2<sup>2</sup>-2, 2<sup>3</sup>-3, 2<sup>4</sup>-4, 2<sup>5</sup>-5, ... , 2<sup>n</sup>-n<br>
1,2,5,12, ... итд<br>
значит последовательность
1) возрастающая <br>
2) неограниченная <br>
3) пятый член - 2<sup>5</sup>-5 = 32-5 = 27


Последовательность b<sub>n</sub>= 1/(1-n), n>1 <br>
Имеет вид 1/(1-2), 1/(1-3), 1/(1-4), 1/(1-5), ... , 1/(1-n)<br>
-1, -1/2, -1/3, -1/4, ... <br>
Значит последовательность<br>
1) убывающая<br>
2) ограниченная, стремится к нулю [-1;0)<br>
3) пятый член: (-1/5)<br>

Последовательность c<sub>n</sub> = -1<sup>n</sup>+√(2n) <br>
(-1<sup>n</sup>)<sup>2</sup> + (√(2n))<sup>2</sup> = -1<sup>2n</sup>+ 2n <br>
Имеет вид -1<sup>2*1</sup>+ 2 * 1,-1<sup>2 * 2</sup>+ 2 * 2, -1<sup>2 * 3</sup>+ 2 * 3, -1<sup>2 * 4</sup>+ 2 * 4, -1<sup>2 * 5</sup>+ 2 * 5, ..., -1<sup>n</sup>+√(2n) <br>
3,5,7,9,11 <br>
Значит последовательность<br>
1) возрастающая<br>
2) неограниченная<br>
3) пятый член = 11

(-1)<sup>2 * 1</sup>+1/(1<sup>2</sup>),(-1)<sup>2 * 2</sup>+1/(2<sup>2</sup>), (-1)<sup>2 * 3</sup>+1/(3<sup>2</sup>), (-1)<sup>2 * 4</sup>+1/(4<sup>2</sup>),(-1)<sup>2 * 5</sup>+1/(5<sup>2</sup>)

Последовательность d<sub>n</sub> = (-1)<sup>2n</sup>+1/(n<sup>2</sup>) <br>
Имеет вид (-1)<sup>2 * 1</sup>+1/(1<sup>2</sup>),(-1)<sup>2 * 2</sup>+1/(2<sup>2</sup>), (-1)<sup>2 * 3</sup>+1/(3<sup>2</sup>), (-1)<sup>2 * 4</sup>+1/(4<sup>2</sup>),(-1)<sup>2 * 5</sup>+1/(5<sup>2</sup>)<br>
1+1, 1+(1/4), 1+(1/9),1+(1/16),1+(1/25), ..., (-1)<sup>2n</sup>+1/(n<sup>2</sup>)<br>
Значит последовательность <br>
1) убывающая<br>
2) ограниченная, стремится к единице (1;2]<br>
3) пятый член = 1+(1/25)<br>

*На языке Python предложить алгоритм вычисляющий численно предел с точностью  ε= 10<sup>-7</sup>

```python
Image(url= "https://i.imgur.com/vNxHe7L.png")
```

```python
def fac(n):
    if n == 0:
        return 1
    return fac(n-1) * n
```

```python
%%time
for n in range(1,171):
    x = (n/((fac(n))**(1/n)))
    print(f'При n={n}, lim(n/(n!^(1/n))) = {"%.7f" % (x)}')
```
