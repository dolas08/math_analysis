Предложить пример функции, не имеющей предела в нуле и в бесконечностях.<br>
lim(f(x))=\= 0 при x стремящемуся к бесконечности? <br>
Ответ: ничего не понятно, какой-то теоритический вопрос, наверно y=cos(x)<br>

```python
def f(x):
    return np.cos(x)
    
x = np.linspace(-10, 10, 200)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

№2<br>
Привести пример функции, не имеющей предела в точке, но определенной в ней.<br>
Ответ:-<br>

```python
def f(x):
    return x**3 - x**2
    
x = np.linspace(-5, 5, 20)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```
